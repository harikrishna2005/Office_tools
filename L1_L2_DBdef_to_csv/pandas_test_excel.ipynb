```python
import openpyxl
import pandas as pd



def get_all_sheets(excel_file_path):
    workbook = openpyxl.load_workbook(excel_file_path)
    # Get all sheet names
    sheet_names = workbook.sheetnames
    return sheet_names


def get_valid_sheets(excel_file_path, index_sheet=None):
    all_sheets = get_all_sheets(excel_file_path)

    # index sheet existence
    if index_sheet not in all_sheets or index_sheet is None:
        raise ValueError(f"{index_sheet} not found in the list. so CANNOT PROCEED.")
        # all_sheets.remove(index_sheet)
    
    all_sheets.remove(index_sheet)
    return all_sheets

def get_tablenames_columnnames_in_dataframe_from_single_sheet(excel_file_path, mysheet_name):
    df_excel = pd.read_excel(excel_file_path, sheet_name=mysheet_name, header=None)

    # =========Get table name details=========
    table_en_logical_name = df_excel[5][4]
    table_jp_logical_name = df_excel[5][3]
    table_physical_name = df_excel[5][1]
    # print(f'{table_jp_logical_name},{table_en_logical_name},{table_physical_name}')



    # =======Get column Details in a data frame ========
    


    numeric_mask = pd.to_numeric(df_excel[0], errors='coerce').notna()
    # column_names = ['col_jp_logical_name','col_en_logical_name','col_physical_name','col_datatype','col_size','col_pk_key','col_description']
    # column_structure_df = df_excel[[1,2,3,5,6,7,14]][numeric_mask]
    column_names = ['col_jp_logical_name','col_en_logical_name','col_physical_name','col_datatype','col_size','col_pk_key']
    column_structure_df = df_excel[[1,2,3,5,6,7]][numeric_mask]


    column_structure_df.columns = column_names

    # =======append table name details to column details ==========
    column_structure_df.insert(0,'table_physical_name',table_physical_name)
    column_structure_df.insert(1,'table_en_logical_name',table_en_logical_name)
    column_structure_df.insert(2,'table_jp_logical_name',table_jp_logical_name)
    column_structure_df['col_ignored']=0

    return column_structure_df
    
def load_df_to_csv(sheet_data_frame, csv_file_path: str, header_csv: bool=False):
    
    # print(csv_file_path)    
    sheet_data_frame.to_csv(csv_file_path, mode='a',encoding='UTF-8', header=header_csv, index=False)

    


def get_ignored_columns(file_path,sheet_individual):
    ignored_color = '808080'
    print('Entered')
    pass
    



    
def start(excel_file_path,index_sheet,csv_path):
    list_of_sheets = get_valid_sheets(excel_file_path,index_sheet)
    
    # INITIALIZE THE CSV FILE
    with open(csv_path, 'w'):
        pass  # This just opens and immediately closes the file, effectively truncating it
    
    # start processing each sheet 
    print('******* STARTED PROCESSING ************')
    first_time_only_header:bool = True
    for mysheet in list_of_sheets:
        # Get ignored columns or fields
        list_of_ignored_columns = get_ignored_columns(excel_file_path,mysheet)


        # Load sheet data in to csv
        sheet_data_frame = get_tablenames_columnnames_in_dataframe_from_single_sheet(excel_file_path,mysheet)
        load_df_to_csv(sheet_data_frame,csv_path,first_time_only_header)

        if first_time_only_header:
            first_time_only_header = False


        print(f'Completed processing for sheet ==>     {mysheet}')
        # for loop ends here
    print(f'\n  Loaded data  :     {csv_path}')
    print("******* PROCESSING COMPLETED ************")
    

# excel_file_path = r"D:\Bluelake_02\REQUIREMENTS\23.DB定義書 (DB Definition)\DB項目定義書（L1_APP_DCS）dcs_v1.0.xlsx"
# index_sheet = 'TABLE_OF_CONTENTS'
# csv_path = r"D:\A7985\Hari\Work\Projects\Blue Lake\VS code\L1_db_table_data.csv"

# excel_file_path = r"D:\Bluelake_02\REQUIREMENTS\23.DB定義書 (DB Definition)\DB項目定義書（L2_APP_DCS）dcs_v1.0.xlsx"
# index_sheet = '目次'
# csv_path = r"D:\A7985\Hari\Work\Projects\Blue Lake\VS code\L2_db_table_data.csv"




excel_file_path = r"D:\A7985\Hari\Work\Projects\Blue Lake\VS code\onedrivefoldercreated\table_generation\DB項目定義書（L2_APP_DCS）dcs_v1.0.xlsx"
index_sheet = '目次'
csv_path = r"D:\A7985\Hari\Work\Projects\Blue Lake\VS code\L1_db_table_data.csv"

start(excel_file_path=excel_file_path,
      index_sheet=index_sheet,
      csv_path=csv_path)

```
